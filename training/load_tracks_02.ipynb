```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# from in2stellar import Tracks, DenseNetworkGrid, set_seed
from interstellar import Tracks, DenseNetworkGrid, set_seed
%matplotlib inline
```

```python
data_path = '/rds/projects/d/daviesgr-cartography/mesa_grid/grid2_zams_agb/simple_grid_alex_ff'
```

```python
tracks = Tracks(data_path)
```

```python
tracks.compile_metadata()
tracks.get_metadata()
```

```python
from interstellar.tracks import DATA_COLS

print(DATA_COLS)
```

```python
cols = DATA_COLS + ['initial_Z', 'surface_he4', 'surface_Z']
print(cols)
```

```python
tracks.compile_data(usecols=cols)
tracks.get_data()
```

Note, CHeB evolutionary stage is included below due to an error in the model data columns. No Core He-burning tracks were computed but some rows are labelled as such. Since evolutionary stage isn't going to be used in the training or models, this is not an issue. 

```python
tracks.get_data('evol_stage').unique()
```

```python
tracks.select_evolution(stages=['dwarf', 'giant', 'CHeB'])['evol_stage'].unique()
```

```python
tracks.select_evolution(stages=['dwarf', 'giant', 'CHeB'], inplace=True)
```

```python
x_cols = ['frac_age', 'initial_mass', 'initial_MLT', 'initial_Yinit', 'initial_Z']
y_cols = ['star_age', 'effective_T', 'radius', 'delta_nu_fit', 'star_feh', 'surface_he4', 'surface_Z']
```

```python
non_log = ['star_feh']
log_cols = [col for col in x_cols+y_cols if not any([col == n for n in non_log])]
log_cols
```

```python
tracks.log10(columns=log_cols, inplace=True)
tracks.get_data()
```

```python
x_cols = [f'log_{col}' if not any([col == n for n in non_log]) else col for col in x_cols ]
y_cols = [f'log_{col}' if not any([col == n for n in non_log]) else col for col in y_cols ]
print(x_cols)
print(y_cols)
```

```python
%matplotlib inline
fig, ax = plt.subplots(figsize=(16, 9))
tracks.get_data().plot(kind='hist', ax=ax,
                       y=x_cols+y_cols,
                       bins=40, alpha=0.5)
```

```python
fac, df = tracks.normalize(loc='50%', columns=x_cols+y_cols)
fac
```

```python
df.head()
```

Let's simplfy the normalization factors by rounding them to 3 decimal places.

```python
precision = 3
loc = round(fac['loc'], 3).to_list()
loc
```

```python
fac, df = tracks.normalize(loc=loc, columns=x_cols+y_cols)
fac
```

```python
df.head()
```

```python
tracks.normalize(loc=loc, columns=x_cols+y_cols, inplace=True)
tracks.get_data().head()
```

```python
x_cols = [f'{col}_norm' for col in x_cols ]
y_cols = [f'{col}_norm' for col in y_cols ]
print(x_cols)
print(y_cols)
```

```python
%matplotlib inline
fig, ax = plt.subplots(figsize=(16, 9))
tracks.get_data().plot(kind='hist', ax=ax,
                       y=x_cols+y_cols,
                       bins=40, alpha=0.5)
```

Reduce size by converting to 32-bit float.

```python
data = tracks.get_data()
data_floats = df.select_dtypes(include=['float'])
data_floats.head()
```

```python
data[data_floats.columns] = data_floats.astype(np.float32)
data.head()
```

To further save space, we can drop the logged cols as these can be easily recovered later.

```python
log_cols
```

```python
data.drop(columns=log_cols, inplace=True)
```

Reset tracks data.

```python
tracks.set_data(data)
```

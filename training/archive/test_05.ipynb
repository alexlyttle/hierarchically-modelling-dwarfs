```python
set_seed(101)
```

```python
train = tracks.sample_data(fraction=0.8, columns=x_cols+y_cols)
train.head()
```

```python
print(f'Length of training data = {len(train)}')
```

```python
test = tracks.get_data(x_cols+y_cols).drop(train.index)
test.head()
```

```python
print(f'Length of test data = {len(test)}')
```

```python
grid = DenseNetworkGrid(path=save_path, name='grid_05', depth=[3, 4, 5], width=64, activation='elu', batch_size=128)
```

```python
grid.make_networks()
```

```python
grid.build_all(len(x_cols), len(y_cols))
```

```python
grid.compile_all(optimizer='nadam', loss='mae', metrics=['mse'])
```

```python
grid.train_all(train, x_cols, y_cols, epochs=1000, validation_split=None,
               callbacks=['checkpoint'], checkpoint_kw={'monitor': 'loss'})
```

```python
import seaborn as sns
```

```python
df_dict = {}
for i, network in enumerate(grid.networks):
#     network = grid.networks[i]
    df_dict[network.name] = np.abs(network.predict(test) - test[y_cols].reset_index(drop=True))
```

```python
df_err = pd.concat(df_dict).reset_index(level=0).rename(columns={'level_0': 'name'})
df_err.head()
```

```python
df_mae = pd.concat(df_dict).mean(axis=1).reset_index(level=0).rename(columns={'level_0': 'name', 0: 'mae'})
df_mae.head()
```

```python
df_mae['log_mae'] = np.log10(df_mae['mae'])
```

```python
order = df_mae.set_index('name').median(level=0).sort_values('log_mae', ascending=False).index
```

```python
# order = 
# fig_xscale = 1.0
# fig, ax = plt.subplots(figsize=(len(order)*fig_xscale, 4))
fig, ax = plt.subplots()
sns.boxplot(data=df_mae, x='name', y='log_mae',
            width=0.3, fliersize=2, color='C0',
            order=order, ax=ax)
# ax.set_yscale('log')
plt.xticks(rotation=45)
plt.yticks(rotation=45);
# fig.set_figwidth(6)
```

```python
df_err_col = df_err.set_index('name').stack().reset_index().rename(columns={'level_1': 'output',
                                                                0: 'abs_err'})

df_err_col['log_abs_err'] = np.log10(df_err_col['abs_err'])
```

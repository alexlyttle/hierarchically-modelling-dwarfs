```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = 'lund2017_table2.txt'

data = np.genfromtxt(path, dtype=str, skip_header=6, skip_footer=1)
```

```python
data.shape
```

```python
colnames = np.array([
    'KIC',
    'numaxT',
    'DnuT',
    'dDnu_dnT',
    'epsilonT',
    'dnu01T',
    'ddnu01_dnT',
    'dnu02T',
    'ddnu02_dnT',
    ])
```

```python
df = pd.DataFrame(data, columns=colnames)
df.head()
```

```python
names = df.columns[1:]
```

```python
for n in names:
    df[f'{n[:-1]}'] = df[f'{n}'].str.split('$').str[0]
    df[f'-e_{n[:-1]}'] = df[f'{n}'].str.split('$').str[1].str.split('_').str[1].str.split('^').str[0].str.strip('{}')
    df[f'+e_{n[:-1]}'] = df[f'{n}'].str.split('$').str[1].str.split('_').str[1].str.split('^').str[1].str.strip('{}')
df.head()
```

```python
df_reduced = df.drop(columns=names)
df_reduced.head()
```

```python
df_reduced.loc[:, df_reduced.columns[1:]] = df_reduced.loc[:, df_reduced.columns[1:]].astype(float)
```

```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import rc
from scipy.stats import beta, uniform
```

```python
# rc("font", family="serif", size=9)
# rc("text", usetex=True)
# rc('text.latex', preamble=r'\usepackage{newtxtext,newtxmath}')
```

```python
x = np.linspace(0.0, 1.0, 500)

plt.figure(figsize=(4.2, 3.4), dpi=300)
plt.plot(x, beta.pdf(x, 1.2, 1.2), label=r'$\mathcal{B}(1.2, 1.2)$')
plt.plot(x, uniform.pdf(x, 0., 1.), label=r'$\mathcal{U}(0.0, 1.0)$')
plt.xlabel('x')
plt.ylabel('probability density')
plt.legend()

plt.savefig('../paper/figures/beta_distribution.png')
```

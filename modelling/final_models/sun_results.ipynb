```python
import numpy as np

from corner import corner
from matplotlib import pyplot as plt
from stellr import Optimizer
```

```python
opt = Optimizer()
opt.load_trace('sun_results/SUN/trace')
```

```python
posterior = opt.trace.posterior
```

```python
var_names = [
    'star/f_evol',
     'star/mass',
     'star/mlt',
     'star/yi',
     'star/zi',
     'star/mhi',
     'star/age',
     'star/teff',
     'star/rad',
     'star/lum',
     'star/dnu',
     'star/mhs'
]
```

```python
n_samples = 20000
```

```python
xs = np.zeros((len(var_names), n_samples))
for i, name in enumerate(var_names):
    xs[i] = posterior[name].values.flatten()
```

```python
labels = [
    r'$f_\mathrm{evol}$',
    r'$M\,(\mathrm{M}_\odot)$',
    r'$\alpha_\mathrm{MLT}$',
    r'$Y_\mathrm{init}$',
    r'$Z_\mathrm{init}$',
    r'$[\mathrm{M/H}]_\mathrm{init}\,(\mathrm{dex})$',
    r'$\tau\,(\mathrm{Gyr})$',
    r'$T_\mathrm{eff}\,(\mathrm{K})$',
    r'$R\,(\mathrm{R}_\odot)$',
    r'$L\,(\mathrm{L}_\odot)$',
    r'$\Delta\nu\,(\mu\mathrm{Hz})$', 
    r'$[\mathrm{M/H}]_\mathrm{surf}\,(\mathrm{dex})$',
]
```

```python
# plt.rc('axes', titlesize=16.)
plt.rc('axes', labelsize=18.)
plt.rc('xtick', labelsize=18.)
plt.rc('ytick', labelsize=18.)

fig = corner(xs.T, labels=labels, labelpad=.15, show_titles=False, quantiles=[.16, .5, .84])
# for ax in fig.get_axes():
#     ax.xaxis.labelpad = 20.
```

```python
fig.savefig('sun_results/SUN/star/corner_plot_2.png', dpi=300, bbox_inches='tight')
```

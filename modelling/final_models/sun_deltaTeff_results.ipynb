```python
import os
import logging
import pandas as pd
import numpy as np
import pymc4 as pm
import tensorflow as tf

from stellr import Optimizer
```

```python
path = 'sun_deltaTeff_results/SUN/trace'

opt = Optimizer()
opt.load_trace(path)
```

```python
posterior = opt.trace.posterior
```

```python
posterior
```

```python
dteff = posterior['star/delta_teff'].values.flatten()
mlt = posterior['star/mlt'].values.flatten()
```

```python
cov_matrix = np.cov(dteff, mlt)
corr_matrix = np.corrcoef(dteff, mlt)
assert cov_matrix[0, 1] == cov_matrix[1, 0]

cov = cov_matrix[0, 1] 
corr = corr_matrix[0, 1]

print(f'Covarience between delta_teff and mlt: {cov:.3f} K (correlation of {corr:.3f})')
```

So this means that delta_Teff correlated positively with alpha_mlt. If we decrease delta_teff - i.e. if our observed temperature scale is underestimated, then we would decrease the best fit alpha_mlt.

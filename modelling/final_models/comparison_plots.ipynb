```python
if not os.path.isdir('comparison'):
    os.mkdir('comparison')
```

```python
plt.style.use('seaborn-paper')
```

```python
s17 = pd.read_csv('../../data/serenelli2017/s17_SDSS.csv')
s17.head()
```

```python
no_pool = pd.read_csv('stars_results/DR14_ASPC/stars/results.csv')
no_pool.head()
```

```python
par_pool = pd.read_csv('population_results/partial_pool/DR14_ASPC/population/stars/results.csv')
par_pool_sun = pd.read_csv('population+sun_results/partial_pool/DR14_ASPC/population/stars/results.csv')
```

```python
max_pool = pd.read_csv('population_results/max_pool/DR14_ASPC/population/stars/results.csv')
max_pool_sun = pd.read_csv('population+sun_results/max_pool/DR14_ASPC/population/stars/results.csv')
```

```python
s17['name'] = 'KIC' + s17['KIC'].astype(str)
s17.head()
```

```python
names = par_pool['name']  # Take names which are in common with all dataframes
```

```python
data = {
    'NP': no_pool.loc[no_pool['name'].isin(names)],
    'PP': par_pool.loc[par_pool['name'].isin(names)],
    'PPS': par_pool_sun[par_pool_sun['name'].isin(names)],
    'MP': max_pool[max_pool['name'].isin(names)],
    'MPS': max_pool_sun[max_pool_sun['name'].isin(names)],
}
```

```python
kde_defaults = dict(linewidth=1, alpha=0.75)
kde_kws = {
    'S17': dict(color='k'),
    'NP': dict(color='C0'),
    'PP': dict(color='C1'),
    'PPS': dict(color='C1', linestyle='--'),
    'MP': dict(color='C2'),
    'MPS': dict(color='C2', linestyle='--')
}
```

```python
delta_ages = {}
for key, val in data.items():
    df = s17.merge(val, on='name')
    delta_ages[f'S17 - {key}'] =  (df['Age'] - df['age_mean'])/df['Age']
```

```python
fig, ax = plt.subplots(figsize=figsize, dpi=300)

for (label, d_age), key in zip(delta_ages.items(), data.keys()):
    sns.distplot(d_age, label=label, hist=False, ax=ax, kde_kws=dict(kde_defaults, **kde_kws[key]))
ax.axvline(color='k', linestyle='--', linewidth=1)
ax.legend()

ax.set_xlabel(r'$\Delta\mathrm{age}\,(\mathrm{Gyr})$')
ax.set_ylabel('KDE')
```

```python
delta_masses = {}
for key, val in data.items():
    df = s17.merge(val, on='name')
    delta_masses[f'S17 - {key}'] =  (df['Mass'] - df['mass_mean'])/df['Mass']
```

```python
fig, ax = plt.subplots(figsize=figsize, dpi=300)

for (label, d_mass), key in zip(delta_masses.items(), data.keys()):
    sns.distplot(d_mass, label=label, hist=False, ax=ax, kde_kws=dict(kde_defaults, **kde_kws[key]))
ax.axvline(color='k', linestyle='--', linewidth=1)
ax.legend()

ax.set_xlabel(r'$\Delta M\,(\mathrm{M}_\odot)$')
ax.set_ylabel('KDE')
```

```python
delta_rads = {}
for key, val in data.items():
    df = s17.merge(val, on='name')
    delta_rads[f'S17 - {key}'] =  (df['Radius'] - df['rad_mean'])
```

```python
fig, ax = plt.subplots(figsize=figsize, dpi=300)

for (label, d_rad), key in zip(delta_rads.items(), data.keys()):
    sns.distplot(d_rad, label=label, hist=False, ax=ax, kde_kws=dict(kde_defaults, **kde_kws[key]))
ax.axvline(color='k', linestyle='--', linewidth=1)
ax.legend()

ax.set_xlabel(r'$\Delta R\,(\mathrm{R}_\odot)$')
ax.set_ylabel('KDE')
```

```python
age_sds = {}
age_sds['S17'] = (s17.loc[s17['name'].isin(names), '+eAge'] - s17.loc[s17['name'].isin(names), '-eAge']) / 2.0 / \
                 s17.loc[s17['name'].isin(names), 'Age'] * 100

for key, val in data.items():
    df = s17.merge(val, on='name')
    age_sds[key] =  df['age_sd'] / df['age_mean'] * 100
```

```python
fig, ax = plt.subplots(figsize=figsize, dpi=300)

for label, age_sd in age_sds.items():
    sns.distplot(age_sd, label=label, hist=False, ax=ax, kde_kws=dict(kde_defaults, **kde_kws[label]))

ax.legend()
ax.set_xlabel(r'$\sigma_\tau / \tau\,(\%)$')
ax.set_ylabel('KDE')
ax.set_xlim(0.0, age_sds['S17'].max())

fig.savefig('comparison/sigma_age.png', bbox_inches='tight')
```

```python
mass_sds = {}
mass_sds['S17'] = (s17.loc[s17['name'].isin(names), '+eMas'] - s17.loc[s17['name'].isin(names), '-eMas']) / 2.0 /\
                 s17.loc[s17['name'].isin(names), 'Mass'] * 100
for key, val in data.items():
    df = s17.merge(val, on='name')
    mass_sds[key] =  df['mass_sd'] / df['mass_mean'] * 100
```

```python
fig, ax = plt.subplots(figsize=figsize, dpi=300)

for label, mass_sd in mass_sds.items():
    sns.distplot(mass_sd, label=label, hist=False, ax=ax, kde_kws=dict(kde_defaults, **kde_kws[label]))

ax.legend()
ax.set_xlabel(r'$\sigma_M / M \, (\%)$')
ax.set_ylabel('KDE')

fig.savefig('comparison/sigma_mass.png', bbox_inches='tight')
```

```python
rad_sds = {}
rad_sds['S17'] = (s17.loc[s17['name'].isin(names), '+eRad'] - s17.loc[s17['name'].isin(names), '-eRad']) / 2.0 /\
                  s17.loc[s17['name'].isin(names), 'Radius'] * 100
for key, val in data.items():
    df = s17.merge(val, on='name')
    rad_sds[key] =  df['rad_sd'] / df['rad_mean'] * 100
```

```python
fig, ax = plt.subplots(figsize=figsize, dpi=300)

for label, rad_sd in rad_sds.items():
    sns.distplot(rad_sd, label=label, hist=False, ax=ax, kde_kws=dict(kde_defaults, **kde_kws[label]))

ax.legend()
ax.set_xlabel(r'$\sigma_R/R \, (\%)$ ')
ax.set_ylabel('KDE')
ax.set_xlim(0.0)
fig.savefig('comparison/sigma_rad.png', bbox_inches='tight')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = '../data/tracks/grid1_Dm0.04_MLT1.9_MS.csv'

grid = pd.read_csv(path)
grid.head()
```

```python
sns.pairplot(grid, vars=['star_age', 'inital_mass', 'initial_feh'])
```

```python
sns.pairplot(grid, vars=['effective_T', 'luminosity', 'delta_nu'])
```

```python
grid = grid.rename(columns=dict(inital_mass='initial_mass'))
```

```python
inputs = ['star_age', 'initial_mass', 'initial_feh']
outputs = ['effective_T', 'luminosity', 'delta_nu']

fig, axes = plt.subplots(2, 3, figsize=(12, 9), dpi=200., sharey=True,
                         gridspec_kw=dict(wspace=0.05))

for i, (in_col, out_col) in enumerate(zip(inputs, outputs)):
    grid.plot(ax=axes[0, i], kind='hist', y=in_col, edgecolor='C0', bins=12, facecolor='none', legend=False)
    grid.plot(ax=axes[1, i], kind='hist', y=out_col, edgecolor='C1', bins=12, facecolor='none', legend=False)
    axes[0, i].set_xlabel(in_col)
    axes[1, i].set_xlabel(out_col)
    
```

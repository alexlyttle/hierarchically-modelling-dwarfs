```python
import numpy as np
import pandas as pd
```

```python
data_path = 'data/serenelli2017_low-mass.csv'

df = pd.read_csv(data_path)
df.head()
```

```python
latex_cols = ['KIC', 'numax', 'enumax', 'Dnu', 'eDnu']
header = [r'\textit{Kepler} ID', r'$\nu_{\max}$', r'$\sigma_{\nu_{\max}}$', r'$\Delta\nu$', r'$\sigma_{\Delta\nu}$']
cols = {l: h for l, h in zip(latex_cols, header)}
print(cols)
# latex_table = df.head().to_latex(columns=latex_cols, header=header, index=False, escape=False)
# print(latex_table)
```

```python
path_to_latex = 'plan/main.tex'
```

```python
from astropy.table import Table
from astropy.io import ascii
```

```python
table = Table.from_pandas(df.rename(columns=cols).head())
table
```

```python
filename = 'plan/tables/serenelli_seismo.tex'
caption = (r'An extract from Table 3 of \citetalias{serenelliFirstAPOKASCCatalog-2017} comprising measurements of ' +
           r'the global asteroseismic parameters from the Sydney pipeline ' +
           r'\citep{huberAutomatedExtractionOscillation-2009} with uncertainties from the spread ' +
           r'of results from several other pipelines.\label{tab:seismo}')
formats = {h: '.3f' for h in header[1:]}
ascii.write(table, output=filename, Writer=ascii.AASTex, include_names=header, formats=formats,
            caption=caption, col_align='rrrrr', overwrite=True)
```

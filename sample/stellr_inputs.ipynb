```python
scale = 'DR14_IRFM'

seis = pd.read_csv('../data/serenelli2017/s17_seis.csv')
lums = pd.read_csv(f'../data/isoclassify/isoclassify_outputs_{scale}.csv')

seis.head()
```

```python
lums.head()
```

```python
lums['KIC'] = lums.id_starname.str.lstrip('KIC').astype(int)
lums.head()
```

```python
# lums['dir_lum_err'] = (lums.dir_lum_err1 - lums.dir_lum_err2) / 2.
lums['dir_lum_err'] = lums[["dir_lum_err1", "dir_lum_err2"]].max(axis=1)
```

```python
data = lums.merge(seis, on='KIC')
data.head()
```

```python
rename_dict = {
    'id_starname': 'name',
    'dir_lum': 'lum',
    'dir_lum_err': 'lum_err',
    'Dnu': 'dnu',
    'eDnu': 'dnu_err',
    'feh': 'mhs',
    'feh_err': 'mhs_err'
}

if 'IRFM' in scale:
    rename_dict['dir_teff'] = 'teff'
#     data['dir_teff_err'] = (data.dir_teff_err1 - data.dir_teff_err2) / 2.
    data['dir_teff_err'] = data[["dir_teff_err1", "dir_teff_err2"]].max(axis=1)
    rename_dict['dir_teff_err'] = 'teff_err'

# Drop columns which share rename values
data.drop(columns=[col for col in rename_dict.values() if col in data.columns], inplace=True)
data.rename(columns=rename_dict, inplace=True)
data.head()
```

```python
cols = [
    'name', 'teff', 'teff_err', 'lum', 'lum_err', 'dnu', 'dnu_err', 'mhs', 'mhs_err', 'logg', 'logg_err'
]

data = data[cols]
data.head()
```

Flag points less than zero or with error greater than 25%

```python
data['bad_data'] = 0
for obs in ['teff', 'lum', 'dnu']:
    print(f'obs={obs}')
    if len(data[data[obs] < 0]) > 0:
        data.loc[data[obs] < 0, 'bad_data'] = 1
        print(data[data[obs] < 0])
    if len(data[data[f'{obs}_err']/data[obs] > 0.25]):
        data.loc[data[f'{obs}_err']/data[obs] > 0.25, 'bad_data'] = 1
        print(data[data[f'{obs}_err']/data[obs] > 0.25])
```

Flag points with obscenely low metallicity or an error of greater than 25% (note since in dex, we multiply by 0.434 to get the log error).

```python
obs = 'mhs'
if len(data[data[obs] < -9]) > 0:
    data.loc[data[obs] < -9, 'bad_data'] = 1
    print(data[data[obs] < -9])
if len(data[data[f'{obs}_err'] > 0.25*0.434]):
    data.loc[data[f'{obs}_err'] > 0.25*0.434, 'bad_data'] = 1
    print(data[data[f'{obs}_err'] > 0.25*0.434])
```

Finally, let's load in S17 masses and flag points outside of the range.

```python
s17 = pd.read_csv('../data/serenelli2017/s17_SDSS.csv')  # Their recommended results
s17.head()
```

```python
s17['name'] = 'KIC' + s17['KIC'].astype(str)
s17.head()
```

```python
df = data.merge(s17, on='name', suffixes=('', '_S17'))
df.head()
```

```python
df['Mass+2e'] = df['Mass'] + 2 * df['+eMas']
df['Mass-2e'] = df['Mass'] + 2 * df['-eMas']  # Note this is negative in the table
df['mhs+2e'] = df['mhs'] + 2 * df['mhs_err']
df['mhs-2e'] = df['mhs'] - 2 * df['mhs_err']
df['logg-2e'] = df['logg'] - 2 * df['logg_err']
```

```python
df['Mass+e'] = df['Mass'] + df['+eMas']
df['Mass-e'] = df['Mass'] + df['-eMas']  # Note this is negative in the table
df['mhs+e'] = df['mhs'] + df['mhs_err']
df['mhs-e'] = df['mhs'] - df['mhs_err']
df['logg-e'] = df['logg'] - df['logg_err']
```

```python
cond = (df['mhs-2e'] > -0.5) & (df['mhs+2e'] < 0.5) & \
    (df['Mass-2e'] > 0.8) & (df['Mass+2e'] < 1.2) & \
    (df['logg-2e'] > 3.8)

df['on_grid_2e'] = 0
df.loc[cond, 'on_grid_2e'] = 1
df.head()
```

```python
cond = (df['mhs-e'] > -0.5) & (df['mhs+e'] < 0.5) & \
    (df['Mass-e'] > 0.8) & (df['Mass+e'] < 1.2) & \
    (df['logg-e'] > 3.8)

df['on_grid_1e'] = 0
df.loc[cond, 'on_grid_1e'] = 1
df.head()
```

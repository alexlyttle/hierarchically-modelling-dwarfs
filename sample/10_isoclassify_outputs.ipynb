```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dr14_aspc = pd.read_csv('../data/isoclassify/isoclassify_outputs_DR14_ASPC.csv')
dr14_aspc.head()
```

```python
dr14_irfm = pd.read_csv('../data/isoclassify/isoclassify_outputs_DR14_IRFM.csv')
dr14_irfm.head()
```

```python
dr16_aspc = pd.read_csv('../data/isoclassify/isoclassify_outputs_DR16_ASPC.csv')
dr16_aspc.head()
```

```python
dr16_irfm = pd.read_csv('../data/isoclassify/isoclassify_outputs_DR16_IRFM.csv')
dr16_irfm.head()
```

```python
dr14 = dr14_aspc.merge(dr14_irfm, on='id_starname', how='outer', suffixes=('_ASPC', '_IRFM'))
dr16 = dr16_aspc.merge(dr16_irfm, on='id_starname', how='outer', suffixes=('_ASPC', '_IRFM'))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

for ax, df, label in zip(axes, [dr14, dr16], ['dr14', 'dr16']):
    df.plot(ax=ax, kind='scatter', x='dir_teff_ASPC', y='dir_teff_IRFM',
                 xerr='dir_teff_err1_ASPC', yerr='dir_teff_err1_IRFM',
                 alpha=0.5, s=4)

    teff_lim = (4500, 7000)
    ax.plot(np.linspace(*teff_lim, 10), np.linspace(*teff_lim, 10), 'k--')
    ax.set_xlim(teff_lim)
    ax.set_ylim(teff_lim)
    ax.set_title(label)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

for ax, df, label in zip(axes, [dr14, dr16], ['dr14', 'dr16']):
    df.plot(ax=ax, kind='scatter', x='dir_lum_ASPC', y='dir_lum_IRFM',
                 xerr='dir_lum_err1_ASPC', yerr='dir_lum_err1_IRFM',
                 alpha=0.5, s=4)
    lum_lim = (0, 15)
    ax.plot(np.linspace(*lum_lim, 10), np.linspace(*lum_lim, 10), 'k--')
    ax.set_xlim(lum_lim)
    ax.set_ylim(lum_lim)
    ax.set_title(label)
```

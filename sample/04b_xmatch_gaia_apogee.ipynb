# *Gaia*-APOGEE cross-match

## A notebook for loading and cross-matching with APOGEE

Cross match with Serenelli - Gaia table

This should work with DR13 through to DR16 (note DR14 and DR15 same for APOGEE).

DR14(15): https://www.sdss.org/dr15/irspec/spectro_data/

DR16: https://www.sdss.org/dr16/irspec/

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.table import Table
```

```python
dr = '14'  # Data release for file nameing

if dr == '16':
    path = '/Users/ajl573/Data/allStarLite-r12-l33.fits'
elif dr == '14' or dr == '15':
    path = '/Users/ajl573/Data/allStar-l31c.2.fits'

with fits.open(path) as hdul:
    data = hdul[1].data
```

```python
data.columns
```

```python
table = Table(data)
table
```

```python
try:
    table.to_pandas()
except ValueError as verr:
    msg = str(verr)

print(msg)
```

```python
drop_cols = eval(msg.split(':')[1])
drop_cols
```

```python
cols = [col for col in table.colnames if col not in drop_cols]
cols
```

```python
df = table[cols].to_pandas()
df.head()
```

Load in the serenelli - gaia cross-match. Kepler has a 2MASS designation ID which may be cross-matched with the APOGEE ID.

```python
df_kg = pd.read_csv('../data/kepler_gaia/s17_gaia_4arcsec.csv')
df_kg.head()
```

```python
df_kg['APOGEE_ID'] = df_kg['tm_designation'].str.replace('2MASS J', '2M')
```

```python
df_kg['gaia_source_id'] = df_kg['source_id']
```

```python
df_output = df_kg[['KIC', 'gaia_source_id', 'nearest_neighbour', 'APOGEE_ID']].merge(df, on='APOGEE_ID')
df_output.head()
```

Due to many-to-many merge where there are duplicates in the APOGEE data, I check some of the flags. I separate the nearest neighbours from the rest so that I can isolate the cases where only APOGEE duplicates.

```python
df_nearest = df_output[(df_output['nearest_neighbour']==1)]
df_other = df_output[~(df_output['nearest_neighbour']==1)]

df_nearest.loc[(df_nearest['APOGEE_ID'].duplicated(keep=False)),
              ['APOGEE_ID', 'TEFF', 'M_H', 'TARGFLAGS', 'SNR']].sort_values('APOGEE_ID')
```

I then do the same check on the remaining non-nearest neighbours. It's still possible that there are Kepler duplicates, but the TARGFLAGS should tell us which is the culprit on the APOGEE side.

```python
df_other.loc[(df_other['APOGEE_ID'].duplicated(keep=False)),
              ['APOGEE_ID', 'TEFF', 'M_H', 'TARGFLAGS', 'SNR']].sort_values('APOGEE_ID')
```

It appears that the duplicates are Kepler Objects of Interest. They also have higher signal to noise. Here we decide to keep the target with the highest SNR.

```python
df_reduced = df_nearest.sort_values('SNR').drop_duplicates('APOGEE_ID', keep='last')  # We keep the high SNR
```

```python
df_final = pd.concat([df_reduced, df_other]).sort_values('KIC')
print('Length = ', len(df_final))
df_final.head()
```

## Save

```python
df_final.to_csv(f'../data/spectroscopy/s17_gaia_4arcsec_apogeeDR{dr}.csv', index=False)
```

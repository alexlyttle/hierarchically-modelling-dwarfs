```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_dr14 = pd.read_csv('../data/spectroscopy/s17_gaia_4arcsec_apogeeDR14.csv')
df_df16 = pd.read_csv('../data/spectroscopy/s17_gaia_4arcsec_apogeeDR16.csv')

df_both = pd.merge(df_dr14, df_df16, on=['KIC', 'nearest_neighbour'], suffixes=('_DR14', '_DR16'))
df = df_both.loc[df_both['nearest_neighbour']==1]
df.head()
```

```python
lim = [4500, 7000]

ax = df.plot(kind='scatter', x='TEFF_DR14', y='TEFF_DR16')
ax.plot(lim, lim, 'k--')
ax.set_xlim(lim)
ax.set_ylim(lim)
```
